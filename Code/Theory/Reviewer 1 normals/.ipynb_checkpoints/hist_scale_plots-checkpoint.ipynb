```python
import matplotlib.pyplot as plt
import scipy.io as sio
import numpy as np
```

```python
data = sio.loadmat('a_stan.mat')
a1_stan = data['a1'].reshape(-1,)
a2_stan = data['a2'].reshape(-1,)
u_stan = data['u'].reshape(-1,)
data = sio.loadmat('a_tanh.mat')
a1_tanh = data['a1'].reshape(-1,)
a2_tanh = data['a2'].reshape(-1,)
u_tanh = data['u'].reshape(-1,)

plt_data = [a1_stan,a2_stan,u_stan,a1_tanh,a2_tanh,u_tanh]
```

```python
def true_disc(x): #[-4,3.75]
    y  = np.zeros((np.shape(x)[0],))
    
    y = 0.2*np.sin(6*x)*np.exp(-1*x)
    y[x>0] = 1 + 0.1*x[x>0]*np.exp(x[x>0]) + 10
    
    return scale*y
scale = 50
lb = np.array(-4.0)  # [-1. 0.]
ub = np.array(3.75) # [1.  0.99]

x_test = np.linspace(lb,ub,1000).reshape(-1,1)
u_true = true_disc(x_test)
```

```python
plt.plot(u_true)
plt.plot(u_tanh)
```

```python
algo_names = ['stan','tanh']
colors = ['r','k']
labels = ['Stan (Proposed)','tanh']

fig,ax = plt.subplots(2,3)


for i in range(2):
    for j in range(2):
        ax[i,j].hist(plt_data[3*i+j],density = True,color = colors[i])
        ax[i,j].set_xlabel('Value',fontsize=12,alpha = 1)
        ax[i,j].set_ylabel('Probability Density',fontsize=12,alpha = 1)
        ax[i,j].tick_params(axis='x', labelsize=8)
        ax[i,j].tick_params(axis='y', labelsize=8)
    ax[i,2].plot(x_test,u_true,'b')
    ax[i,2].plot(x_test,plt_data[3*i+2],'r--')
    ax[i,2].set_xlabel('$x$',fontsize=12,alpha = 1,math_fontfamily='cm')
    ax[i,2].set_ylabel('$u$',fontsize=12,alpha = 1, math_fontfamily='cm')
    
for i in range(2):
    for j in range(2):
        ax[i,j].set_title(labels[i]+' Layer '+ str(j+1),fontsize=14, math_fontfamily='cm')   
        
    ax[i,2].set_title(labels[i]+ ' Prediction')
    # ax[i].legend(prop={'size':8}, loc = 'upper right')
    # # ax_loss.legend(prop={'size':6})
    # ax[i].set_xlabel('Epochs',fontsize=12,alpha = 1)
    # ax[i].set_ylabel('Loss',fontsize=12, math_fontfamily='cm')
    # ax[i].tick_params(axis='x', labelsize=8)
    # ax[i].tick_params(axis='y', labelsize=8)
    # # ax[i].set_yscale('log')
    # ax[i].set_xlim([0,num_steps])
# i = i+1
    
# plt_titles = ["Low","Medium","High"]   
# for i in range(3):
#     axs[0,i].set_title(plt_titles[i],fontsize=14, math_fontfamily='cm')    
    
# plot_rows = ['(a)','(b)','(c)']

# for i in range(3):
#     axs[i,0].text(-0.5,0.5,plot_rows[i],fontsize=14,transform=axs[i,0].transAxes)

fig.subplots_adjust(left=-1,bottom = -0.7)
plt.savefig('layer_scaling.pdf', format='pdf',pad_inches=0, bbox_inches='tight')
```
