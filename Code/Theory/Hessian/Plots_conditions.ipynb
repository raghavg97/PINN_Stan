```python
algos = ['stan','tanh','atanh','swish','rowdy']
algo_labels = ['Stan (Proposed)','tanh','tanh(AF)','Swish','Rowdy']

layer_num = 1

max_steps = 40

colors = ['r','k','b--','g:','c-.']

fig,ax = plt.subplots(1,1)



for i in range(4):
    dat = scipy.io.loadmat('Condition_'+algos[i]+'.mat')
    k =  dat['k'][:max_steps]


    ax.plot(range(0,max_steps),k,colors[i],label = algo_labels[i])
    ax.set_xlim([0,max_steps])
    ax.figure.set_size_inches(5,5)
    ax.set_yscale('log')



ax.set_xlabel('Epochs',fontsize=12,alpha = 1)
ax.set_ylabel('SCN',fontsize=16,alpha = 1)
ax.legend()   
# ax.set_title("Layer "+str(layer_num+1),fontsize=18)
        
# fig.subplots_adjust(top = 2.0)   
fig.savefig('SCN.pdf', format='pdf', bbox_inches='tight')
```
