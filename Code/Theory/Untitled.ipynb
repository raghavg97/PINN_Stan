```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4,4,100)

sig = np.tanh(x)
sig1 = 1-np.square(sig)
sig2 = -2*sig*sig1

x_ax = 0*x;
```

```python
fig,ax = plt.subplots()

ax.plot(x,x_ax,'k-.')
ax.plot(x_ax,x,'k-.')
ax.plot(x,sig,'b',label = 'tanh',linewidth = 3)
ax.plot(x,sig1,'r',label = '1st Derivative',linewidth = 3)
ax.plot(x,sig2,'m',label = '2nd Derivative',linewidth = 3)

ax.legend(loc = 2,fontsize = 14)

ax.set_xlim([-4,4])
ax.set_ylim([-1.1,1.1])

plt.savefig('tanh_derivs.svg', format='svg',pad_inches=0, bbox_inches='tight',transparent=True)
```

```python
s = 0.3
np.random.seed(1234)
sig = np.random.normal(0,s,(1000,))
sig[sig>1] =1
sig[sig<-1] =-1

sig1 = np.random.normal(1-s**2,np.sqrt(2*s**4),(1000,))
# sig1[sig1>1] = 2-sig1[sig1>1]
sig1[sig1>1] = 1

sig2 = np.random.normal(0,np.sqrt(4*s**2*(15*s**4 - 6*s**2 + 1)),(1000,))
sig2[sig2>1] =1
sig2[sig2<-1] =-1

fig,ax = plt.subplots()

ax.hist(sig,color = 'b',bins = 25,alpha = 0.7,label = 'tanh',density = 'True')
ax.hist(sig1,color = 'r',bins = 25,alpha = 0.9,label = '1st Derivative',density = 'True')
ax.hist(sig2,color = 'g',bins = 25,alpha = 0.5,label = '2nd Derivative',density = 'True')
ax.legend()
plt.savefig('tanh_derivs_dist.svg', format='svg',pad_inches=0, bbox_inches='tight',transparent=True)
```
