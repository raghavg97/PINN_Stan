```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def stan(x,beta):
    y = np.tanh(x) + beta*x*np.tanh(x)
    
    return y

def stan_der(x,beta):
    y = (1-np.square(np.tanh(x))) + beta*x*(1-np.square(np.tanh(x))) + beta*np.tanh(x)
    
    return y
```

```python
x = np.linspace(-2,2,500).reshape(-1,1)
y1 = stan(x,0.5)
y2 = stan(x,1)
y3 = stan(x,1.5)
ytanh = stan(x,0)

y4 = np.zeros((500,1))

y = np.linspace(-1,4,500).reshape(-1,1)
x4 = np.zeros((500,1))
```

```python
x = np.linspace(-4,4,500).reshape(-1,1)
y_der_1 = stan_der(x,0.5)
y_der_2 = stan_der(x,1)
y_der_3 = stan_der(x,1.5)
y_der_tanh = stan_der(x,0)

y_der_4 = np.zeros((500,1))

y = np.linspace(-2,3,500).reshape(-1,1)
x_der_4 = np.zeros((500,1))
```

```python
fig,ax = plt.subplots(1,2)
fig.dpi=300
#fig.size = (96,48)
fig.tight_layout()

ax[0].plot(x,ytanh,'m',linewidth = 2,label = '$\\beta = 0$ (tanh)')
ax[0].plot(x,y1,'r',linewidth = 2,label = '$\\beta = 0.5$')
ax[0].plot(x,y2,'b',linewidth = 2,label = '$\\beta = 1.0$')
ax[0].plot(x,y3,'g',linewidth = 2,label = '$\\beta = 1.5$')
ax[0].plot(x,y4,'k-.')
ax[0].plot(x4,y,'k-.')


ax[0].set_xlim([-2,2])
ax[0].set_ylim([-1,2])
ax[0].legend(prop={'size':8})
ax[0].set_xlabel('$x$', fontsize=12, math_fontfamily='cm')
ax[0].set_ylabel('$Stan(x)$',fontsize=12, math_fontfamily='cm')
# plt.savefig('Stan.eps', format='eps',pad_inches=0, bbox_inches='tight')
# plt.savefig('Stan.svg', format='svg',pad_inches=0, bbox_inches='tight')
ax[0].set_aspect(1)

#-----------------------------------------------------------------------------------------

ax[1].plot(x,y_der_tanh,'m',linewidth = 2,label = '$\\beta = 0$ (tanh)')
ax[1].plot(x,y_der_1,'r',linewidth = 2,label = '$\\beta = 0.5$')
ax[1].plot(x,y_der_2,'b',linewidth = 2,label = '$\\beta = 1.0$')
ax[1].plot(x,y_der_3,'g',linewidth = 2,label = '$\\beta = 1.5$')

ax[1].plot(x,y_der_4,'k-.')
ax[1].plot(x_der_4,y,'k-.')


ax[1].set_xlim([-4,4])
ax[1].set_ylim([-2,2.5])
ax[1].legend(prop={'size':8},loc = 2)
ax[1].set_xlabel('$x$', fontsize=12, math_fontfamily='cm')
ax[1].set_ylabel('$Stan$ gradient',fontsize=12, math_fontfamily='cm')
ax[1].set_aspect(1.3333)

ax[0].tick_params(axis='x', labelsize=8,length = 1)
ax[0].tick_params(axis='y', labelsize=8,length = 1)
ax[1].tick_params(axis='x', labelsize=8,length = 1)
ax[1].tick_params(axis='y', labelsize=8,length = 1)


# plt.savefig('Stan gradient.eps', format='eps',pad_inches=0, bbox_inches='tight')
#plt.savefig('Stan gradient.svg', format='svg',pad_inches=0, bbox_inches='tight')
ax[0].text(1.5,-0.8,'(a)',fontsize=12)
ax[1].text(3,-1.70,'(b)',fontsize=12)

#plt.savefig('Stan.pdf', format='pdf',pad_inches=0, bbox_inches='tight')
plt.savefig('Stan.svg', format='svg',pad_inches=0, bbox_inches='tight',transparent = True)
```
